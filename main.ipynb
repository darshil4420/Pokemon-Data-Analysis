```python
import pandas as pd
data=pd.read_excel('Dataset.xlsx')
data.head()
```

```python
#Rename column for better usage
data.rename(columns={'Sp. Atk': 'Sp._Atk', 'Sp. Def': 'Sp._Def','Type 1':'Type1','Type 2':'Type2'}, inplace=True)
data.head()
```

```python
data.isnull().sum(axis=0)  #Null Values
```

```python

#Available Categorise in Type2
list(data['Type2'].value_counts().index)
```

```python
#We are replacing HP with it's mean
data['HP'].fillna(data['HP'].mean(), inplace=True)
data.head(5)
```

```python
data.isnull().sum(axis=0)  #Null Values
```

```python
#1.  Selecting Pokemon Having HP between 40 and 75

Pokemon_HP = data[(data['HP']>40) & (data['HP']<=75)]
```

```python
#3	Fill null values with mean of column HP, Attack and Defense.
#We are replacing HP with it's mean
data['HP'].fillna(data['HP'].mean(), inplace=True)
data['Attack'].fillna(data['Attack'].mean(),inplace=True)
data['Defense'].fillna(data['Defense'].mean(),inplace=True)
```
